```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np
np.random.seed(0)
```

```python
iris=load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df['species']=pd.Categorical.from_codes(iris.target,iris.target_names)
df
```

```python
df['is_train']=np.random.uniform(0,1,len(df))<=.75
df.head()
```

```python
train,test=df[df['is_train']==True],df[df['is_train']==False]
print('number of observations in the training data:',len(train))
print('number of observations in the test data:',len(test))
```

```python
features=df.columns[:4]
features
```

```python
y=pd.factorize(train['species'])[0]
y
```

```python
clf=RandomForestClassifier(n_jobs=2,random_state=0)
clf.fit(train[features],y)
```

```python
clf.predict(test[features])
```

```python
clf.predict_proba(test[features])[0:10]
```

```python
preds=iris.target_names[clf.predict(test[features])]
preds[0:5]
```

```python
pd.crosstab(test['species'],preds,rownames=['Actual Species'],colnames=['Predicted Species'])
```

```python
accuracy=clf.score(test[features],pd.factorize(test['species'])[0])
print('Accuracy:',accuracy*100)
```

```python
from lightgbm import LGBMClassifier
```

```python
lgbm=LGBMClassifier(n_estimators=100,learning_rate=0.1,random_state=0)
lgbm.fit(train[features],y)
```

```python
pred=lgbm.predict(test[features])
pred
```

```python
pred_species=iris.target_names[pred]
pred_species
```

```python
pd.crosstab(test['species'],pred_species,rownames=['Actual Species'],colnames=['Predicted Species'])
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(test['species'],pred_species)
print('Accuracy:',accuracy)
print('Accuracy(%)',accuracy*100)
```
